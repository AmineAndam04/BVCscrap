```python
#You must respect the notation
load.notation()
```

```python
# get data of a single stock market
data=load.loadata('BCP',start='2021-09-01',end='2021-09-10')
data
```

```python
#Get the data of more than one stock
data=load.loadmany('BCP','BMCI',start="2021-08-30",end='2021-09-04')
data
```

```python
#Get intraday data 
#Exemple 01
data=load.getIntraday('BCP')
data
```

```python
#Get intraday data 
#Exemple 02
data=load.getIntraday('CIH')
data
```
